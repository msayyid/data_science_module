Data pre-processing involves transforming raw data into a clean, organized format suitable for analysis or machine learning models. While the other listed techniques directly modify or structure the data itself, Data Visualization is a technique used to graphically represent data (via charts, graphs, and maps) to communicate insights and identify patterns. ## 8 - A
Data discretization is a data reduction technique specifically designed to transform continuous numeric data into a finite set of intervals or categorical "bins". By grouping a large range of numeric values (e.g., ages 0–100) into smaller, meaningful categories (e.g., "Youth," "Adult," "Senior"), it simplifies the dataset and reduces its complexity while maintaining the underlying patterns. This process is essential for many data mining and machine learning algorithms, such as Decision Trees and Naïve Bayes, which often perform better with discrete input rather than continuous variables.
