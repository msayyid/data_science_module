```python
# useful functions to create numpy arrays
a = np.linspace(0, 1, 11) # 11 elements, uniformly spaced, 0 to 1
print(a, a.shape)

b = np.arange(0, 20, 2.5) # 0 to 20 elements spaced by 2.5
print(b, b.shape)

c = np.zeros(5) # 1d array consisting of 5 zeros
print(c.shape)
print(c)

M = np.zeros((2, 5)) # 2x5 (2 rows, 5 columns) array of zeros
print(M.shape)
print(M)

w = np.ones(7) # 1d array consisting of 7 ones
print(w.shape)
print(w)

N = np.ones((3, 2)) # 3x2 (3 rows, 2 columns) of ones
print(N.shape)
print(N)

I = np.eye(4) # 4x4 array with ones on the diagonal and zeros elsewhere
print(I.shape)
print(I)

R = np.random.rand(2, 4) # 2x4 array of random values from a uniform distribution
print(R.shape)
print(R)
```

```python
# # Pandas
# Pandas is a Python library for working with data in tables.
# A DataFrame is like a spreadsheet (rows and columns).
# A Series is one column from that table.
# Unlike NumPy arrays (usually one data type), Pandas can have different data types in 
# different columns (for example: names as text, ages as numbers).

import numpy as np
import pandas as pd
print("we're good")
```

```python
s = pd.Series([1, 2, 4, 9, 16, 25])
# pandas automatically labels each item starting from zero
# s is a single column of data
df = pd.DataFrame(np.random.randn(10, 10)) # make me a table with 10 rows and 10 columns filled with random numbers

print("The series s is\n", s)
print("\n the dataframe df is\n", df)
print("the first two rows of df are\n", df.head(3))
print("the last two rows of the df are:\n", df.tail(2))
```

```python
# When you display a Pandas Series or DataFrame, you see index numbers on the left (for rows) 
# and at the top (for columns).
# By default, these are integers like 0, 1, 2…
# But you can set your own row labels using index and your own column names using columns.


df = pd.DataFrame(np.random.randint(1, 100, [4, 4]), index=["r1", "r2", "r3", "r4"], columns=["c1", "c2", "c3", "c4"])
print("the dataframe df:\n", df)
print("\nthe row labels (index)\n", df.index)
print("\nthe column labels (columns)\n", df.columns)
```
