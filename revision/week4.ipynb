# Machine Learning - Quick Notes

## 1. What is Machine Learning

Machine Learning (ML) is a **field of computer science and artificial intelligence that allows computers to learn patterns from data and make predictions or decisions without being explicitly programmed.**

Key ideas:

* Uses **algorithms**
* Learns **patterns from data**
* Improves with **experience (data)**

---

# Types of Machine Learning

## 1. Supervised Learning

Learning from **labeled data**.

The dataset contains **inputs and correct outputs (labels)**.

Example:

| Input          | Label |
| -------------- | ----- |
| Email text     | Spam  |
| House features | Price |

Main tasks:

* **Classification** - predict categories
  Example: spam / not spam
* **Regression** - predict continuous values
  Example: house price

Examples:

* Sentiment analysis (positive / negative)
* Rain prediction
* House price prediction

---

## 2. Unsupervised Learning

Learning from **unlabeled data**.

The algorithm tries to **discover patterns or structure** in the data.

Common task:

* **Clustering** - grouping similar data points

Examples:

* Customer segmentation
* Genetic similarity grouping
* Market basket analysis

---

## 3. Semi-supervised Learning

Uses **both labeled and unlabeled data**.

Usually:

* Small amount of labeled data
* Large amount of unlabeled data

The model may **assign labels to unlabeled data (pseudo-labeling)**.

---

# Important Machine Learning Concepts

## Training Data

Data used to **train the model**.

The algorithm learns patterns from this dataset.

---

## Test Data

Data used to **evaluate the model performance** on unseen data.

---

## Dataset Splitting

A dataset is usually split into:

* **Training set**
* **Test set**

Each class/label should be **properly represented in both sets**.

This is often done using **stratified sampling**.

---

# Supervised Learning Tasks

## Classification

Used when the **output variable is a category**.

Examples:

* Red / Blue
* Spam / Not spam
* Positive / Negative review
* Rain / No rain

---

## Regression

Used when the **output variable is a continuous numeric value**.

Examples:

* House price
* Temperature
* Weight
* Sales amount

---

# Unsupervised Learning Tasks

## Clustering

Grouping **similar data points together** based on features.

Examples:

* Customer segmentation
* Document grouping
* Genetic grouping

Distance measures often used:

* **Euclidean Distance**

---

# Euclidean Distance

Measures the **closeness between two data points**.

Often used in:

* Clustering
* K-Nearest Neighbors (KNN)

---

# Naive Bayes

Naive Bayes assumes:

* **Features are independent** of each other.

Example:
Features:

* Color
* Shape
* Size

The algorithm assumes these features **do not depend on each other**.

Used mainly for **classification problems**.

Examples:

* Spam detection
* Sentiment analysis
* Document classification

---

# Linear Regression and Outliers

Linear Regression is **sensitive to outliers**.

Reason:
It minimizes **squared errors**, so large errors from outliers strongly affect the regression line.

---

# Good Machine Learning Methods Should Be

* **Accurate**
* **Fast**
* **Scalable**

# Machine Learning - 1 Minute Cheat Sheet

**Supervised Learning**

* Data **has labels**
* Model learns **input → output**
* Tasks:

* Data **has no labels**
* Model **finds patterns or groups**
* Main method:
