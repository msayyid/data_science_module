### Data integration

* **Integration**: combining data from multiple sources into one storage/dataset.

### Data visualization outcomes

* Visualization gives:

  * **Meaningful representation** (shows patterns/trends)
  * **User-friendly representation** (easy to understand)
* MCQ answer: **Both a and b**.

### Data preprocessing vs visualization

* **Data preprocessing**: preparing/cleaning data before analysis (cleaning, discretization, aggregation, normalization).
* **Data visualization**: presenting data for understanding/communication.
* MCQ answer (not a preprocessing method): **Data visualization**.

### Data cleaning tasks

* Data cleaning includes:

  * **Removing noisy data**
  * **Correcting inconsistencies**
  * **Transforming data into usable format**
* MCQ answer: **All of the above**.

### Data transformation

* **Normalization** (scaling values) and **aggregation** (summarizing values) are both transformation processes.
* MCQ answer: **True**.

### Binning and noise

* **Binning**: grouping numeric values into intervals (bins) to smooth/reduce noise.
* MCQ answer: **True**.

### Discretization

* **Discretization**: converting continuous numeric data into categories/bins.
* Most useful for: **Numeric data**.

### Attribute naming across sources

* Same attribute can have different names in different sources and must be standardized when integrating.
* MCQ answer: **True**.

### Min-Max normalization

* **Min-Max normalization** rescales values linearly (often to 0–1):
  [
  x' = \frac{x - x_{min}}{x_{max} - x_{min}}
  ]
* MCQ answer: **True** (linear transformation).

### JPEG compression

* **JPEG is lossy**: it reduces file size by permanently discarding some information.
* MCQ answer: **True**.

### Redundancy detection in data reduction

* **Dimension reduction** helps detect/remove redundant features (features carrying similar information).
* MCQ answer: **Dimension reduction**.

## Data cleaning needed in the given table example

### Problems

* **Height column**: mixed units/formats (`160.4cm`, `158cm`, `5 feet 3 inches`).
* **Payroll column**: mixed currencies (`£`, `$`), symbols/commas, and a likely error/outlier (`$360,0000`).
* **Departments column**: same department written differently (`&` vs `and`, `CSE` vs full name).
* **Missing data**: two empty rows and some missing fields.

### Fixes

* Convert all **heights** to one unit (cm) and store as numeric.
* Standardize **payroll** to one currency, remove symbols/commas, validate/fix extreme or typo values.
* Standardize **department names** using a mapping list.
* Remove empty rows; handle missing values (fill or drop depending on analysis).
