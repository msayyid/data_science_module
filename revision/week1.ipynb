````python
import numpy as np
import statistics

data = np.array([4.2, 4.3, 4.0, 4.1])

# Mean
mean = np.mean(data)

# Median
median = np.median(data)

# Population variance
variance_population = np.var(data)

# Sample variance
variance_sample = np.var(data, ddof=1)

# Mode using statistics
try:
    mode = statistics.mode(data)
except statistics.StatisticsError:
    mode = "No unique mode"

# standard deviation 
standard_dev = 0.012
variance = standard_dev ** 2

print(f"Variance from std={standard_dev} is {variance}")

print("Mean:", mean)
print("Median:", median)
print("Population Variance:", variance_population)
print("Sample Variance:", variance_sample)
print("Mode:", mode)
````

# Statistics and Data Concepts Summary

## 1. Qualitative vs Quantitative Attributes

**Qualitative (categorical):**

* Nominal
* Ordinal
* Binary

These describe categories, not numbers.

**Quantitative (numerical):**

* Integer
* Interval scaled
* Ratio scaled

These represent measurable numerical values.

---

## 2. Population vs Sample

**Population**
The entire collection of data.

**Sample**
A subset taken from the population.

If a question says "whole dataset" → population.
If data represents part of a larger group → sample.

---

## 3. Mean

Average of values.

Formula:

Mean = sum(x) / N

NumPy:

```python
np.mean(data)
```

---

## 4. Median

Middle value after sorting.

NumPy:

```python
np.median(data)
```

---

## 5. Mode

Most frequently occurring value.

If no value repeats → no unique mode.

Using statistics:

```python
import statistics
statistics.mode(data)
```

---

## 6. Range

Measures total spread.

Range = max − min

NumPy:

```python
np.ptp(data)
```

---

## 7. Variance

Measures how spread out values are from the mean.

### Population Variance

Divide by N.

Formula:

σ² = Σ(x − μ)² / N

NumPy:

```python
np.var(data)
```

### Sample Variance

Divide by N − 1 (Bessel’s correction).

Formula:

s² = Σ(x − x̄)² / (N − 1)

NumPy:

```python
np.var(data, ddof=1)
```

Rule:

* Whole dataset → population
* Estimating larger group → sample

---

## 8. Standard Deviation

Square root of variance.

More interpretable because it has same units as data.

### Population:

```python
np.std(data)
```

### Sample:

```python
np.std(data, ddof=1)
```

Relationship:

Variance = (Standard Deviation)²
