# Linear Regression Notes

## 1. Regression

**Regression** is a technique used to **predict continuous numeric values** based on input variables.

Example:

* Predict **house price**
* Predict **weight from height**
* Predict **sales from store size**

---

# Types of Regression

## Simple Linear Regression

Uses **one predictor variable**.

[
y = b_0 + b_1x
]

* **y** - predicted value (dependent variable)
* **x** - input variable (independent variable)
* **b₀** - intercept
* **b₁** - slope

---

## Multiple Regression

Uses **more than one predictor variable**.

Example:

[
y = b_0 + b_1x_1 + b_2x_2
]

---

# Regression Equation Components

### Intercept (b₀)

Value of **y when x = 0**.

Also called **y-intercept**.

---

### Slope (b₁)

Shows **how much y changes when x increases by 1**.

---

# Dependent vs Independent Variables

### Dependent Variable (y)

The **value we want to predict**.

Example:

* house price
* weight
* sales

---

### Independent Variables (x)

The **input features used to predict y**.

Example:

* height
* size of store
* number of rooms

---

# Supervised Learning

**Linear Regression is a supervised learning algorithm.**

Why?

Because the training data contains:

* **input (x)**
* **correct output (y)**

---

# Model Evaluation Metrics

## Mean Squared Error (MSE)

Measures the **average squared difference between predicted and actual values**.

[
MSE = \frac{1}{n}\sum (y - \hat{y})^2
]

Where

* **y** = actual value
* **ŷ** = predicted value

Rule:

**Lower MSE = better model**

---

## Root Mean Squared Error (RMSE)

Square root of MSE.

[
RMSE = \sqrt{MSE}
]

Also used to measure **prediction error**.

Lower RMSE = better model.

---

# Least Squares Method

Linear regression finds the **best fit line** using **Least Squares**.

It minimizes the **sum of squared errors** between predicted and real values.

---

# Overfitting

**Overfitting happens when a model learns noise instead of the real pattern.**

Signs:

* Very complex model
* Fits training data perfectly
* Performs poorly on new data

Overfitting is **more likely when:**

* dataset is **small**
* model is **too complex**

---

# Outliers

**Outliers are extreme data points that are far away from other values.**

Example:

```
10, 11, 12, 13, 100
```

Here **100** is an outlier.

Outliers can **affect regression models**.

---

# Linear Regression Characteristics

In Linear Regression:

* **Dependent variable is continuous**
* **Independent variables can be continuous or discrete**
* Relationship is **linear (straight line)**

---

# Polynomial Regression

Used when the relationship between variables is **non-linear**.

Example equation:

[
y = b_0 + b_1x + b_2x^2
]

Allows the model to **fit curved patterns** in data.

---

# Regression vs Classification

### Regression

Predicts **continuous values**

Examples:

* price
* temperature
* weight

---

### Classification

Predicts **categories**

Examples:

* spam / not spam
* high / low rating
* customer churn yes / no

---

# Model Selection

Sometimes **multiple candidate models** are created.

We choose the **best model using evaluation metrics** like:

* MSE
* RMSE

Model with **lower error** is better.
