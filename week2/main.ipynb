```python
# matrix operations

# np.array() converts Python data (lists or tuples) into a NumPy array object.
# NumPy performs element-wise addition.
# It adds each element in the same position.

# `np.array()` → creates a mathematical matrix  
# `+` and `-` → element-wise operations  
# `np.dot()` → true matrix multiplication 

x = np.array( ((2, 3), (3, 5) ))
# [ 2   3 ]
# [ 3   5 ]
y = np.array( ((1, 2), (5, -1)))
# [ 1   2 ]
# [ 5  -1 ]

# Each one is a 2×2 matrix.
# What is a MATRIX MULTIPLICATION
# You multiply a row from the first matrix
# with a column from the second matrix
# and add the results.


addition = x + y
print("x:", x)
print("y:", y)
print("addition of x and y is", addition)

subtraction = x - y
print("difference between x and y", subtraction)

# [ 2   3 ]
# [ 3   5 ]

# [ 1   2 ]
# [ 5  -1 ]

# Row 1 of x → [2, 3]
# Column 1 of y → [1, 5]

# Now multiply matching positions:

# (2 × 1) + (3 × 5) = 17 - the first value in the result matrix [17  ?] [?   ?]
# (2 * 2) + (3 * -1) = 1 - the next value [17   1] [?  ?]

# Row 2 of x → [3, 5]
# Column 1 of y → [1, 5]
# (3 * 1) + (5 * 5) = 28 - [17  1] [28  ?]
# (3 * 2) + (5 * -1) = 1 => [17 1] [28 1]

# Matrix multiplication means:
# Take a row from X
# Take a column from Y
# Multiply matching elements
# Add them
# That gives one output value
# Repeat for all columns of Y
# Then move to next row of X
dot_product = np.dot(x, y)
print("dot product of x and y", dot_product)
```

```python
# matrix() function of numpy
# for np.matrix objects
# '*' means matrix multiplication, in here: x * y is equivalent to np.dot(x, y)
# BUT for np.array, x * y is element wise multiplication
x = np.matrix( ((2, 3), (3, 5)))
y = np.matrix( ((1, 2), (5, -1)))
print(x)
print(y)

multiplication = x*y
print("multiplicaiton of x and y:\n", multiplication)

# best practice do not use np.matrix(), use array and x @ y; @ is clean matrix multiplication operator

x = np.array(((2, 3), (3, 5)))
y = np.array(((1, 2), (5, -1)))
print("the modern way using @ operator for matrix multiplication:\n", x @ y)
```

```python
# Generating Binominal Distribution
"""
binominal distribution
it models the number of successes in a fixed number of independent trials.
it is used when: 
- you repeat an experiment n times; 
- each trial has only two outcomes (success or failure)
- the probability success p stays the same
ex: tossing a coin 10 times and counting how many heads apper
"""
# n - number of trials
# p - probability of success in each trial
X = stats.binom(10, 0.2) # n = 10; p = 0.2 - 20% chance of success each time
# X - REPRESENTS THE NUMBER OF SUCCESSES IN n TRIALS, WHERE EACH TRIAL HAS PROBABILITY p OF SUCCESS.
"""
PMF (probability mass function) - the probability of getting exactly k successes
CDF (cumulative distribution function) - the probability of getting k or fewer successes
Mean = n * p; the mean is the average value you expect if you repeat the experiment many times
Variance = n*p*(1-p) - it tells how spread out the results are from the mean,
    it measures how much the outcomes fluctute;
    if variance is small - results stay close to the mean
    if variance is large - results are more spread out
Standard deviation (std) - std = sqrt(variance)
    it measures spread like variance, but it is in the same units as the data and it is easier to understand

mean = 2 -> average number of successes is 2
std = 1.26 -> most results are roughly withing 1.26 of 2; so most outcomes will be 2 +- 1.26
"""
print(X.rvs()) # number of successes in n time trial where success chance is 20%
print(X.pmf(3)) # P(X=3)probability of getting exactly 3 successes out of 10
print(X.cdf(4)) # P(X <= 4) probability of getting at most 4 successes
print(X.mean())
print(X.var())
print(X.std())
```

```python
# Generating normal distribution
# Mean → center of curve
# Variance → spread
# Standard deviation → square root of variance
# PDF → height of curve
# CDF → actual probability up to a point
A = stats.norm(3, mt.sqrt(16)) # mean = 3, standard deviation = 4, where variance is 16
print(A.pdf(4))
print(A.cdf(2))
```
