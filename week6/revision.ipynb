### Probability

**Sample space (Ω)**

* The set of *all possible outcomes* of an experiment.
* Example: rolling a die -> Ω = {1, 2, 3, 4, 5, 6}

**Event**

* A *subset* of the sample space (some outcomes you care about).
* Example: “roll an even number” -> {2, 4, 6}

**Conditional probability**

* Probability that event **A** happens *given that* event **B** already happened.
* Notation: **P(A | B)**
* Intuition: you are “zooming in” to only the cases where **B** is true.

**Random variable (RV)**

* A function that maps outcomes to numbers.
* Example: roll two dice, let X = sum. Outcomes are pairs (1..6, 1..6), but X is a number from 2 to 12.

**Probability distribution**

* A rule that tells you how probabilities are assigned.
* For discrete RVs: probability for each value.
* For continuous RVs: probability “density” across values.

**PMF (Probability Mass Function)**

* Used for **discrete** random variables.
* Gives **P(X = x)** directly (exact value probabilities).

**PDF (Probability Density Function)**

* Used for **continuous** random variables.
* PDF value is *not* a probability by itself.
* Probabilities come from *area under the curve* over an interval:

  * **P(a ≤ X ≤ b)** = area under PDF from a to b

**CDF (Cumulative Distribution Function)**

* Works for **both discrete and continuous**.
* Definition: **F(x) = P(X ≤ x)** (probability that X is at most x).
* CDF is always non-decreasing and goes from 0 to 1.

**Normal (Gaussian) distribution**

* A very common continuous distribution with a “bell curve” shape.
* Defined by:

  * **mean (μ)**: center of the curve
  * **standard deviation (σ)**: spread (how wide it is)
* Many real-world measurements are approximately normal.

### Linear Algebra

**Matrix**

* A rectangular table of numbers (rows x columns).
* Used to represent data, linear transformations, systems of equations, etc.

**Identity matrix (I)**

* A square matrix with 1s on the main diagonal and 0s elsewhere.
* Acts like “1” for matrix multiplication:

  * **A · I = A** and **I · A = A** (when sizes match)

**Matrix addition**

* You can add matrices only if they have the same shape.
* Add element-by-element:

  * (A + B)[i, j] = A[i, j] + B[i, j]

**Matrix multiplication**

* Not element-by-element (that’s different).
* If A is (m x n) and B is (n x p), then A·B is (m x p).
* Each entry is a dot product:

  * (A·B)[i, j] = sum over k of A[i, k] * B[k, j]
* Order matters: **A·B ≠ B·A** in general.
