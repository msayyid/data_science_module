```python
import numpy as np
```

```python
HP_Var = np.loadtxt("HP.csv", dtype=int)
print(HP_Var)
```

```python
# calculating the mean of HP
# The mean is the average of a set of numbers.
mean_of_HP = np.mean(HP_Var)
print(mean_of_HP)
```

```python
# Question 1
# calculating the median of HP
# The median is the middle value in a set of numbers when they are arranged in order.
median_of_HP = np.median(HP_Var)
print(median_of_HP)
```

```python
# Question 2
# calculate the median of mpg file
mp_var = np.loadtxt("MpG.csv", dtype=float)
print(mp_var)

mean_of_mpg = np.mean(mp_var)
print("mean of mpg is ", mean_of_mpg)

median_of_mp = np.median(mp_var)
print(median_of_mp)
```

```python
# importing SciPy library
from scipy import stats

# The mode is the value that appears most frequently in a dataset.

mode_of_hp = stats.mode(HP_Var)
print(mode_of_hp)
```

```python
# Question 3
# calculate the mode of ctype
ctype_var = np.loadtxt("CType.csv", dtype=int)
print(ctype_var)

mode_of_ctype = stats.mode(ctype_var)
print(mode_of_ctype)

# calculate the variance of HP
# Variance measures how spread out your data is
# it tells you how far the numbers are from teh mean on average
# variance is small - if all numbers are close to the mean
# variance is large - if numbers are far from the mean
variance_of_hp = np.var(HP_Var)
print(HP_Var)
print(variance_of_hp)
```

```python
# Question 4
# calculate the variance of mpg

variance_of_mpg = np.var(mp_var)
print("variance of mpg: ", variance_of_mpg)
```

```python
# calculating the standard deviation of HP
# Standard deviation = average distance from the mean (in normal units).

# Standard deviation tells you how much the data typically differs from the mean.
# It is simply the square root of variance.
s_deviation_of_hp = np.std(HP_Var)
print(s_deviation_of_hp)
```

```python
# Question 5
# calcualate the standard deviation of mpg
s_deviation_of_mpg = np.std(mp_var)
print(mp_var)
print(s_deviation_of_mpg)
```

```python
# calculating the range of quantity
# Range is the difference between the largest and smallest values in a dataset.
# range = max value - min value

# method 1 using max and min function
print("the data itself:\n", HP_Var)
largest = max(HP_Var)
print("maximum = ", largest)
smallest = min(HP_Var)
print("minimum = ", smallest)
range = largest - smallest
print("Range= ", range)


# method 2 using for loop
def range(x):
    max = 0
    min = 999999999
    
    for value in x:
        if value > max:
            max = value
        if value < min:
            min = value
    range = max - min
    return range
range_value = range(HP_Var)
print(range_value)
```

```python
# question 6
# calculate range of mpg using both methods

# method 1
print(mp_var)
largest = max(mp_var)
smallest = min(mp_var)
range_value = largest - smallest
print("range of mpg: ", range_value)

def find_range(data):
    max_val = data[0]
    min_val = data[0]

    for value in data:
        if value > max_val:
            max_val = value
        if value < min_val:
            min_val = value
    
    return max_val - min_val

print("range using for loop: ", find_range(mp_var))
```
