## Data as a point in a space

Think of each row in a dataset as a **point** on a graph.

Example (2 features):

* **x-axis:** Age
* **y-axis:** Salary

Depending on your goal, you can choose either one as the **predictor (input X)** and the other as the **label (output y)**:

* Predict **Salary** from **Age** (Age = predictor, Salary = label)
* Predict **Age** from **Salary** (Salary = predictor, Age = label)

Both are valid as *prediction* tasks. The choice depends on what you want to predict.

---

## Predictors and Labels: Association vs Causation

Sometimes people incorrectly read a prediction model like it proves **causation**:

* “Predictor causes the label.”

But a model usually works because of **association** (a relationship in the data), not because one variable truly causes the other.

Two attributes can look associated for several reasons:

* **Direct causation:** one actually influences the other

  * Example: more experience may increase salary
* **Common cause:** a third factor affects both

  * Example: education affects both age (years in study) and salary
* **Sampling/selection effects:** the way data was collected creates a pattern

  * Example: dataset only includes full-time workers, making age and salary look more linked than they are

**Key rule:**

* All **causation** implies **association**.
* But not all **association** implies **causation**.
