```python
# We’re going to make a NumPy array called x_r that looks like a row: 
# it has 1 row and 3 columns (so it’s a row vector).

# Then we’ll use np.transpose to flip it, turning it into x_c, which looks 
# like a column: it has 3 rows and 1 column (so it’s a column vector).

import numpy as np
```

```python
x_r = np.array([0, 1, 2], ndmin=2)
# ndmin=2 means make sure array has at least two dimensions
print("Vector x_r is:\n", x_r)
print("the shape of x_r is: ", x_r.shape)

x_c = np.transpose(x_r)
# transpose swaps rows and columns
print("\nVector x_c is:\n", x_c)
print("the shape of x_c is: ", x_c.shape)

# (1, 3) = row vector
# (3, 1) = column vector
# (3,) = just a 1D array (no real row/column structure)
# That last one is important - 1D arrays do not truly behave like row or column matrices.


# Note that we could have used `x_c = np.array([[0], [1], [2]])` to produce a column vector 
# directly, instead of producing a row vector and then using transposition.
```

```python
x_c = np.array([[0], [1], [2]]) # create 1 column vector
print("Vector x_c is:\n", x_c)
```

```python
# Create a design matrix by adding a column of 1s (for the intercept)
# next to x_c, so we can use it in linear regression (y = β0 + β1x)

X = np.column_stack([np.ones(x_c.shape), x_c])
print("The design matrix X is:\n", X)
print("\nthe shape of the matrix X is:", X.shape)
```

```python
# 2. Regression: Linear Regression
# Let's explore a simple regression problem. First, we will define and plot our 
# training dataset, which consists of 10 samples described by one predictor `x` and one label `y`.

import matplotlib.pyplot as plt
```

```python
x = np.array([0.3000, -0.7700, 0.9000, -0.0400, 0.7400, -0.5800, -0.9200, -0.2100, -0.5400, 0.6800], ndmin=2).T
y = np.array([1.1492,  0.3582, 1.9013,  0.9487, 1.3096,  0.9646,  0.1079,  1.1262,  0.6131, 1.0951], ndmin=2).T
# Create column vectors for x and y and plot the training data points
plt.plot(x, y, 'o', label="Training Data")
plt.xlabel("x (attribute)", fontsize=18)
plt.ylabel("y (label)", fontsize=18)
plt.xlim(-1,1) 
plt.ylim(-0.5,2.5)
plt.grid(alpha=0.2)
plt.legend(fontsize=12)
plt.show()
```
